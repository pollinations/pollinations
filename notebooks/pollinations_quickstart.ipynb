Generate images, text, and audio with the free Pollinations API. No API key needed!

## Image Generation

```
import requests
from IPython.display import Image, display

prompt = "a cute robot painting a landscape, digital art"
url = f"https://gen.pollinations.ai/image/{requests.utils.quote(prompt)}?model=flux&width=512&height=512&nologo=true"

response = requests.get(url)
display(Image(data=response.content))
print(f"Generated image for: {prompt}")
```

## Text Generation (OpenAI-compatible)

```
import requests, json

response = requests.post(
    "https://gen.pollinations.ai/v1/chat/completions",
    json={
        "model": "openai",
        "messages": [{"role": "user", "content": "Write a haiku about AI art"}]
    }
)

result = response.json()
print(result["choices"][0]["message"]["content"])
```

## Audio / Text-to-Speech

```
import requests
from IPython.display import Audio, display

text = "Hello! I was generated by the Pollinations API."
url = f"https://gen.pollinations.ai/audio/{requests.utils.quote(text)}?voice=nova"

response = requests.get(url)
with open("speech.mp3", "wb") as f:
    f.write(response.content)

display(Audio("speech.mp3"))
print(f"Generated speech: {text}")
```

## Using the Python SDK

```
import pollinations as ai
from IPython.display import Image, display

# Generate an image
image = ai.image("a serene japanese garden at sunset", model="flux", width=512, height=512)
image.save("garden.png")
display(Image(filename="garden.png"))

# Generate text
response = ai.text("Explain quantum computing in one sentence")
print(response)
```

## Available Models

Check available models dynamically:

```
import requests, json

# Text models
models = requests.get("https://gen.pollinations.ai/v1/models").json()
print("Text models:")
for m in models["data"][:10]:
    print(f"  - {m['id']}")
print(f"  ... and {len(models['data']) - 10} more\n")

# Image models
img_models = requests.get("https://gen.pollinations.ai/image/models").json()
print("Image models:")
for m in img_models:
    print(f"  - {m['name']}")
```
